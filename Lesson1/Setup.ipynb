# 3. A primer on learning anything

So that we can learn effectively, there some things we have to keep in mind.

These have helped me throughout my life, and I think it makes a lot of sense in terms of **learning** anything.

## 3.1. A drowning man can't swim

<img src="https://howtobeastoic.files.wordpress.com/2016/09/image1.jpeg">

When we're learning its important to know our **circle of competence**.
Our circle of competence is defined by how much we know at the moment on a certain topic.

A **misalignment** of your circle of competence with your **own expectations** or **other expectations** causes some trouble.

It is important to be **honest** to ourselves about what we know, and go **little by little**.

This will allow you to have **strong foundations**, so that you can build on top of them.


If you **overestimate** yourself, there will be a point in time where there's is going to be a **LOT** of new information.
By this point you will be *"drowning"* in information, and will be unable to make any **sense** out of it!

## 3.2. Intellectual Honesty - there are no stupid questions

<img src="https://i0.wp.com/wp.production.patheos.com/blogs/exploringourmatrix/files/2013/07/wpid-Photo-Jul-21-2013-1024-PM.jpg">

In order to learn effectively, it is **crucial** that we have **intellectual honesty**. What is this about? 

It simply means that everybody in this room is intelligent and is able to **think**. 
Having our **circle of competence** in mind will help us realise if we need help or not.
If you have **questions** its because you are thinking, so questions and **doubts** you have shouldn't go **under the radar**.


All questions should be asked, to me, your peers, or anyone. No question is a stupid question.
