## 6.4. Indexing strings - Importance 6/10

````python
astring = "Hello world!"
print(astring[3:7])
````

This prints a **slice** of the string, starting at index 3, and ending at index 6. But why 6 and not 7? 

**Again, most programming languages do this - it makes doing math inside those brackets easier.**

1. If you just have one number in the brackets, it will give you the single character at that index. 
2. If you leave out the first number but keep the colon, it will give you a slice from the start to the number you left in. (`[:7]`)
3. If you leave out the second number, if will give you a slice from the first number to the end. (`[3:]`)

You can even put negative numbers inside the brackets. They are an easy way of starting at the end of the string instead of the beginning. This way, -3 means "3rd character from the end".

## 6.5. Reverse a string - Importance 11/10

This is the easiest way of reversing a string (and also a **list**!!).

It is a cool little trick that can save you a lot of time.

This happens because of the **slicing** rules.

The regular **slicing** syntax goes as follows `alist[begin:end]`, so it will get you the elements of the list between indexes `begin` and `end`.
If you omit the `begin`, ending up with `alist[:end]` will get you all the elements of the list from the beggining to the `end` index.
Analogously, if you omit `end`, the same will happen.

Here the syntax for this trick has an extra step.
You can also have `alist[begin:end:step]`. This step will tell how many elements to "skip".

````python
alist = [1,2,3,4,5]
alist[::-1]
````

Since the **step** is 1, it will get us all elements.

````python
alist[::2]
````

So if we have a **step** with value -1, we will have the reversing effect.

````python
astring = "Hello world!"
print(astring[::-1])
````

## 8.3. "While" loops

`while` loops repeat as long as a certain boolean condition is met. 

For example:

````python
# Prints out 0,1,2,3,4

count = 0
while count < 5:
    print(count)
    count += 1  # This is the same as count = count + 1
````

## 8.4. "break" and "continue" statements

`break` is used to exit a for loop or a while loop

`continue` is used to skip the current block, and return to the `for` or `while` statement. 

A few examples:



````python
# Prints out 0,1,2,3,4

count = 0
while True:
    print(count)
    count += 1
    if count >= 5:
        break

print()

# Prints out only odd numbers - 1,3,5,7,9
for x in range(10):
    # Check if x is even
    if x % 2 == 0:
        continue
    print(x)
````

## 9.1. What are Functions?
Functions are a convenient way to divide your code into useful blocks

1. order our code

## 9.2. How do you write functions in Python?

A block is a area of code of written in the format of:

```
def name_of_function(argument1, argument2):
    1st block line
    2nd block line
    ...
```
Functions in python are defined using the keyword `def`, followed with the **function's name**. 

For example:

````python
def my_function():
    print("Hello From My Function!")
my_function()
````

**Functions** receive **arguments**. Arguments are like the `x` for your $f(x)$.

For example:



````python
def my_function_with_args(username, greeting):
    print("Hello, %s , From My Function!, I wish you %s"%(username, greeting))
my_function_with_args('MEEC', 'all the best')
````

Functions may return a value to the caller, using the keyword `return`.

For example:



````python
def sum_two_numbers(a, b):
    return a + b
sum_two_numbers(1, 1)
````

## 9.3. How do you call functions in Python?

Simply write the function's name followed by `()`, placing any required **arguments** within the brackets. 

For example, lets call the functions written above (in the previous example):

````python
# Define our 3 functions
def my_function():
    print("Hello From My Function!")

def my_function_with_args(username, greeting):
    print("Hello, %s , From My Function!, I wish you %s"%(username, greeting))

def sum_two_numbers(a, b):
    return a + b

# print(a simple greeting)
my_function()

#prints - "Hello, John Doe, From My Function!, I wish you a great year!"
my_function_with_args("John Doe", "a great year!")

# after this line x will hold the value 3!
x = sum_two_numbers(1,2)
````

## 9.4. Optional arguments

**Optional** arguments are arguments that take on a **default** value.

In this example the default value is 10 (`y=10`). 

````python
def add(x, y=10):
    return x + y
print(add(1))
print(add(1, y=1))
````

You can **call** the function without specifying the **optional** argument (`add(1)`), which will make `y` take on the **default** value of 10.

Each little doll (or **function**) has its own **scope** (the things she can look at).

A *Matryoshka* doll (or **function**) can only look inside itself.
So, it doesn't know how many dolls are **outside** her, just **inside** her.

````python
eg1 = [1,2,3,4,5]
````

In the below function we are appending an element to the declared list inside the function. 

The `eg2` variable declared inside the function is a **local** variable.

## 9.6. Lambda Functions

**Lambda** functions are called **anonymous** functions.

Why are they anonymous? Because they have no name!

They are used in cases where you don't want to do a function body for simple tasks.

For example, multiplying 2 numbers shouldn't require a fully fledged function:

````python
multiply = lambda x, y: x*y
````

````python
multiply(2, 3)
````
